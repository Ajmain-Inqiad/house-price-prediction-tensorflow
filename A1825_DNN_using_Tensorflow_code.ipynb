## Importing dataset

Here we are importing the dataset

```python
# reading train data
dataset= pd.read_csv("train.csv", header = 0, delimiter = ",")

# removing the Id column from dataframe as it has no impact on house price
dataset.drop('Id',axis = 1, inplace = True)

# getting the numerical features
intdataset = dataset.select_dtypes(exclude=['object'])

# getting the object features
stringdataset = dataset.select_dtypes(include=['object'])

# creating empty dataframe to save encoded value of object features
objectdataset = pd.DataFrame()

# loop to convert categorical data to numerical format. As it is a regression problem, i am converting all feature 
# value into numrecial format. So that it is easy for further data processing, visualization and calculation
for i in stringdataset.columns:
# as i need simple convertion from categorical data to numerical data, i am going to use pandas factorize function
    integer_encoded = pd.factorize(stringdataset[i])[0] 
    #factorize() method helps to get the numeric representation of an array by identifying distinct values.
    objectdataset[i] = integer_encoded # saving the encoded data into dataframe
    
dataset = pd.concat([objectdataset, intdataset], axis=1, sort=False) 
# joining numerical features data and encoded object features data into main dataframe
dataset = dataset.reindex(sorted(dataset.columns), axis=1) # sorting the column
dataset = dataset.fillna(0) #filling all missing value with 0
dataset.head() # printing first 5 row of dataframe
```

## Removing Outliers

```python
#graphical respenstation of all features using boxplot
plt.rcParams["figure.figsize"] = [24,30]
dataset.plot(kind='box', subplots=True, layout=(12,7), sharex=False, sharey=False)
plt.show()
```

*To detect the outliers, i have used boxplot which are shown in above graphs. As we can see from the above diagrams we have a few outliers. we need to handle them as they might create problem while training our dataset and lead to wrong answer.*

To remove outliers I have used IsolationForest from sci-kit learn library. IsolationForest is an algorithm with a low linear time complexity and a small memory requirement without employing any distance or density measure. It builds a good performing model with a small number of trees using small sub-samples of fixed size, regardless of the size of a data set. 

Every perception is given an abnormality score and the accompanying choice can be made on its premise: 

1. Score near 1 demonstrates oddities 

2. Score a lot littler than 0.5 demonstrates typical perceptions 

3. In the event that all scores are near 0.5 than the whole example does not appear to have obviously particular irregularities

Here i have used default value of IsolationForest except max_samples and random state. I have given max_samples value 100  if max_samples is larger than the number of samples provided, all samples will be used for all trees that means no sampling. so to avoid that, i have given less value than total sample in max_samples. In random_state i have used value 42 because this is used by the random number generator. So if random_state is set, random_state is the random number generator value.

```python
#Generating the model
model = IsolationForest(max_samples = 100, random_state = 42)

#Fitting the dataset into model
model.fit(dataset)

#getting prediction value from dataset
y_predict_outliers = model.predict(dataset)

#converting into dataframe
y_predict_outliers = pd.DataFrame(y_predict_outliers, columns = ['High'])

#getting the outliers from dataset
dataset = dataset.iloc[y_predict_outliers[y_predict_outliers['High'] == 1].index.values]

#removing the outliers from dataset
dataset.reset_index(drop = True, inplace = True)

#printing the number of outlier it has found and shape of dataset after removing the outlier
print("Number of Outliers:", y_predict_outliers[y_predict_outliers['High'] == -1].shape[0])
print("Number of rows without outliers:", dataset.shape[0])
```

## Scaling data and standardization

Here i am standardization the dataset. It is important to standardize the dataset. By using this process, the mean of all the input features equal to 0 and also converts their variance to 1. Using this system, it ensures that there is no bias while training the model due to the different scales/values of all input features. Without this, neural network might get confused and give a higher weight to those features which have a higher average value than others.

To standardize data, I have used MinMaxScaler Method as it is sensitive to outliers and feature value falls within a bounded interval.

First I am converting dataframe to numpy matrix for easier calculation process and then scaling that matrix and saving that into dataset

```python
#Getting all the column name of dataset
total_train_columns = list(dataset.columns)

#Getting all the features column name of dataset
total_features = list(dataset.columns)

#removing label from features column
total_features.remove('SalePrice')

#converting dataset into numpy matrix
train_matrix = np.matrix(dataset)

#converting features value of dataset into numpy matrix and also dropping the label column
train_features_matrix = np.matrix(dataset.drop('SalePrice',axis = 1))

#converting label column to numpy array by reshape
train_label = np.array(dataset.SalePrice).reshape((dataset.shape[0],1))

#preprocessing starts

#initiate preprocessing for label column and fitting the column
preprocessing_label = MinMaxScaler()
preprocessing_label.fit(train_label)

#initiate preprocessing for whole dataset and fitting that dataset
preprocessing_train_matrix = MinMaxScaler()
preprocessing_train_matrix.fit(train_matrix)

#initiate preprocessing for features columns and fitting those 
preprocessing_train_features_matrix = MinMaxScaler()
preprocessing_train_features_matrix.fit(train_features_matrix)

#saving all the transformed value in dataset
dataset = pd.DataFrame(preprocessing_train_matrix.transform(train_matrix),columns = total_train_columns)
```

Note that in the above dataframe when I print the dataset, they no longer have values in different scales but rather they all are in the same scale.

## Split the dataset into train and test set

To split the dataset, I have used **train_test_split** method from sci-kit learn library.

```python
#getting label column name of dataset
label = "SalePrice"

#getting the label column values from dataset
prediction_set = dataset.SalePrice

#spliting the dataset into 67% train data and 33% test data
x_train, x_test, y_train, y_test = train_test_split(dataset[total_features] , prediction_set, test_size=0.33, random_state=42)

#converting test label column into list
y_test = y_test.values

#converting train label column into list
y_train = y_train.values
```

## Building Deep Neural Network using Tensorflow

### Input, Hidden & Output Layers

The model consist of three layers: Input, Hidden and Output. 

The **Input Layers** are the input features and the **Output Layers** are output label.

Between **Input Layer** and **Output Layer** there are 3 **hidden layers**. The **1st hidden layer** contains **512 neurons**. Following hidden layers are half the size of the previous layer, which means **2nd hidden layer** contains **256 neurons** and **3rd hidden layer** has **128 neurons**. After 3 hidden layers there is **output layer**.

```python
#number of features column of dataset
number_features = x_train.shape[1]

# Number of Neurons
number_neurons_1 = 512
number_neurons_2 = 256
number_neurons_3 = 128

neuralnetwork = tf.InteractiveSession()
```

Placeholder are essential for neural network. There two placeholders in order to fit the model: X contains the networkâ€™s inputs (all the input features) and Y the networkâ€™s output. The shape of the placeholders correspond to (None, n_features). The None argument shows that yet we do not yet know the number of observations that flow through the neural network in each batch, because of that we keep it flexible.

```python
#Setting the placeholder input and output
X = tf.placeholder(dtype=tf.float32, shape=(None, number_features))
Y = tf.placeholder(dtype=tf.float32, shape=(None))
```

To initialize a neural network, it is important to keep the scale of the input variance constant, so that it does not explode by reaching the final layer. Here I have used two default initializer strategies.

In variance_scaling_initializer method, I have used **fan_avg** mode which means average number of inputs and output connections and **uniform distributon**

```python
#Initializing the neural network
sigma = 1
weight_initializer = tf.variance_scaling_initializer(mode="fan_avg", distribution="uniform", scale=sigma)
bias_initializer = tf.zeros_initializer()
```

As in this model, Multilayer Perceptron(MLP) is used, the second dimension of the previous layer is the first dimension in the current layer for weight matrices. It means each layer is passing its output as input to the next layer. The biases dimension equals the second dimension of the current layerâ€™s weight matrix, which equals the number of neurons in this layer.

```python
# Hidden weights:

#Layer 1: Variables for hidden weights and biases
W_hidden_1 = tf.Variable(weight_initializer([number_features, number_neurons_1]))
bias_hidden_1 = tf.Variable(bias_initializer([number_neurons_1]))

#Layer 2: Variables for hidden weights and biases
W_hidden_2 = tf.Variable(weight_initializer([number_neurons_1, number_neurons_2]))
bias_hidden_2 = tf.Variable(bias_initializer([number_neurons_2]))

#Layer 3: Variables for hidden weights and biases
W_hidden_3 = tf.Variable(weight_initializer([number_neurons_2, number_neurons_3]))
bias_hidden_3 = tf.Variable(bias_initializer([number_neurons_3]))

# Output weights:
#Output layer: Variables for output weights and biases
W_out = tf.Variable(weight_initializer([number_neurons_3, 1]))
bias_out = tf.Variable(bias_initializer([1]))
#end of defining the network


#start of architecture of neural network
#hidden layers of the network are transformed by activation functions. So for that, ReLU(Rectified Linear Unit) is used.
#Because ReLU is sparsity and reduced likelihood of vanishing gradient.
#On the other hand, Sigmoid or others might generate some non-zero value resulting in dense representations
hidden_1 = tf.nn.relu(tf.add(tf.matmul(X, W_hidden_1), bias_hidden_1))
hidden_2 = tf.nn.relu(tf.add(tf.matmul(hidden_1, W_hidden_2), bias_hidden_2))
hidden_3 = tf.nn.relu(tf.add(tf.matmul(hidden_2, W_hidden_3), bias_hidden_3))

out = tf.transpose(tf.add(tf.matmul(hidden_3, W_out), bias_out))
```

Cost Function is used to generate a measure of deviation between the networkâ€™s prediction and the actual target. Here I have used Mean Squared Error to compute the average squared deviation between predictions and targets.

Optimizer Function is used to adapt the networkâ€™s weight and bias variables during training. I have used GradientDescentOptimizer as it is easier to use and it first computes gradient and then make a jump in that direction amplified.

```python
# Cost function
mse = tf.reduce_mean(tf.squared_difference(out, Y))
# Optimizer
lr=0.1
opt = tf.train.GradientDescentOptimizer(learning_rate=lr).minimize(mse)
```

Now it is time to fit the neural network. We have already initialize all the required placeholders, variables, initializers, cost functions and optimizers of the network.

At first, I have run the **global_variables_initializer** function to initialize the declared variables and functions. I have used batch_size to feed the network in batch. The training dataset gets divided into (n/batch_size) batches that are sequentially fed into the network by placeholder.

A batch_size data flows through network until it reaches the output layer where it gets computes with actual result. After that tensorFlow conducts an optimization step and updates the networks parameters, equals to the selected learning scheme. After getting updated, the next batch is ready to process through that updated network. One full iteration over all batches is called an epoch. Here I am running 20 epochs.

```python
# Run initializer
neuralnetwork.run(tf.global_variables_initializer())

#Fitting the neural network
batch_size = 100
epochs = 10
cost_history = np.empty(shape=[1],dtype=float)
#running till epochs
for e in range(epochs):
    for i in range(0, len(y_train) // batch_size):
        start = i * batch_size #getting the start point of dataset
        batch_x = x_train[start:start + batch_size] #getting the feature column values from start to batch size
        batch_y = y_train[start:start + batch_size] #getting the target column values from start to batch size
        neuralnetwork.run(opt, feed_dict={X: batch_x, Y: batch_y}) #feeding the network
    cost_history = np.append(cost_history, neuralnetwork.run(mse, feed_dict={X: x_train, Y: y_train}))
    print("For Epoch ", e+1 ," Mean Squared Error: ",neuralnetwork.run(mse, feed_dict={X: x_train, Y: y_train}))

#printing learning cost
plt.rcParams["figure.figsize"] = [5,4]
plt.plot(range(len(cost_history)),cost_history)
plt.axis([0,epochs,0,np.max(cost_history)])
plt.ylabel("Cost(MSE)")
plt.xlabel("Epoch")
plt.show()


pred = neuralnetwork.run(out, feed_dict={X: x_test}) #feeding the test set to get the predicted value
rmse = neuralnetwork.run(tf.sqrt(tf.reduce_mean(tf.square(tf.subtract(pred, y_test))))) #calculating the final RMSE score
#printing the RMSE score
print("Root Mean Square Error: {0:0.3f}".format(rmse))
```

As we can see from the result that, Mean Squared Error after every epoch is reducing.

## Conclusion

### Why Deep Neural network works better than other algorithms
<br>
<p style="text-align: justify">
    DNN outperforms Random Forest in many cases. Deep networks are used to learn and discover inherent/generic features in objects that will be helpful in their classification. Random Forest is an ensemble approach based on decision trees to learn different objects. In Random Forest, it is assumed that those features are already available from some mechanism. </p>

<p style="text-align: justify">
    SVM's size is fixed. They are parametric models. In DNN we have a lot of hidden layers with sizes h<sub>1</sub> through h<sub>n</sub> depending on the number of features, plus bias parameters, and those make up the model. Besides Multi-layer perceptron (like in DNN) is able to find relation between features. DNN may have any number of outputs, while support vector machines have only one. For example, if there is a model to build to receognize digits, each support vector machine would recognize exactly one digit, and fail to recognize all others. Since each handwritten digit cannot be meant to hold more information than just its class, it makes no sense to try to solve this with an deep neural network. But here, these features are all inter-related, deep neural network regression makes more sense than support vector machine regression.
    </p>

<p style="text-align: justify">
    Deep neural network can learn arbitrary boundary, while decision trees only detect boundary like rectangle. Besides, decision tree is not good while the model is trained on batch; as data is continously coming and we need to modify our model based on that. DNN is capable of reflecting the information of new instance on a model very efficiently by just changing the weight values.</p>
